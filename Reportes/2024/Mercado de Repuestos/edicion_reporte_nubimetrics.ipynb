```python
import pandas as pd
repuestos_2024 = pd.read_csv('/Users/martincarrasco/Desktop/Martín_Carrasco/Reportes/2024/Mercado de Repuestos/Ventas mercado de repuestos 2024.csv')
```

```python
repuestos_2024 = repuestos_2024.iloc[1:]
repuestos_2024
```

```python
repuestos_2024.rename(columns = {'Titulo_Publicacion': 'Producto', 'level3': 'Categoría', 'level4': 'Subcategoría', 'Periodo.Mes': 'Mes', 'Nombre_Vendedor': 'Seller', 'Volumen_de_Ventas_Moneda_Local': 'Ventas'}, inplace = True)
repuestos_2024
```

```python
texto_cols = ['Producto', 'Marca', 'Categoría', 'Subcategoría']

repuestos_2024[texto_cols] = repuestos_2024[texto_cols].astype(str)
repuestos_2024['Mes'] = repuestos_2024['Mes'].astype(int)
repuestos_2024['Unidades_Vendidas'] = repuestos_2024['Unidades_Vendidas'].replace(' ', '')
repuestos_2024['Ventas'] = repuestos_2024['Ventas'].replace(' ', '')
repuestos_2024.info()
```

```python
import re

repuestos_2024['Unidades_Vendidas'] = repuestos_2024['Unidades_Vendidas'].str.replace(r'\s+', '', regex=True)

repuestos_2024['Unidades_Vendidas'] = pd.to_numeric(repuestos_2024['Unidades_Vendidas'], errors='raise')
repuestos_2024.info()
```

```python
repuestos_2024['Ventas'] = repuestos_2024['Ventas'].str.replace(r'\s+', '', regex=True)

repuestos_2024['Ventas'] = pd.to_numeric(repuestos_2024['Ventas'], errors='raise')
repuestos_2024.info()
```

```python
repuestos_2024['Año'] = 2024

cols = list(repuestos_2024.columns)

cols.remove('Año')
mes_index = cols.index('Mes')
cols.insert(mes_index, 'Año')

repuestos_2024 = repuestos_2024[cols]

repuestos_2024
```

```python
import calendar

def obtener_ultimo_dia(año, mes):
    ultimo_dia = calendar.monthrange(año, mes)[1]
    return pd.Timestamp(year=año, month=mes, day=ultimo_dia)

repuestos_2024['Fecha_Compra'] = repuestos_2024.apply(
    lambda row: obtener_ultimo_dia(row['Año'], row['Mes']),
    axis=1
)

repuestos_2024['Fecha_Compra'] = repuestos_2024['Fecha_Compra'].dt.strftime('%Y-%m-%d')

cols = list(repuestos_2024.columns)

cols.remove('Fecha_Compra')

mes_index = cols.index('Mes')
cols.insert(mes_index + 1, 'Fecha_Compra')

repuestos_2024 = repuestos_2024[cols]

repuestos_2024
```
