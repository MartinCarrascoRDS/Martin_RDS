```python
# Separar el tamaño del neumático de la descripción
patron = r"(\d{3}[ /]?\d{2}[ /]?[rR]?/?\d{2})"
```

```python
neumaticos['Medida_raw'] = neumaticos["Titulo_Publicacion"].str.extract(patron, expand=False)
neumaticos['Medida_raw'] = neumaticos['Medida_raw'].fillna("No encontrada")
neumaticos
```

```python
neumaticos_ne = neumaticos[neumaticos['Medida_raw'] == "No encontrada"]
neumaticos_ne
```

```python
# Estandarizar medidas
def estandarizar_medida(medida):
    if pd.isna(medida):
        return "No encontrada"
    
    # Extraer los números con regex
    match = re.match(r"(\d{3})[ /]?(\d{2})[ /]?[rR]?/?(\d{2})", medida)
    if match:
        ancho, perfil, aro = match.groups()
        return f"{ancho}/{perfil}R{aro}"
    else:
        return "No encontrada"

# Aplicar estandarización
neumaticos["Medida"] = neumaticos["Medida_raw"].apply(estandarizar_medida)

# Mostrar resultados
print(neumaticos[["Titulo_Publicacion", "Medida"]])
```
