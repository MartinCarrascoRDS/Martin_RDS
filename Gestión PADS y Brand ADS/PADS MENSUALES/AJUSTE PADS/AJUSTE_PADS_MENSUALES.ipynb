# LIMPIEZA DE BASE DE DATOS DE PADS

```python
import pandas as pd
import numpy as np
```

```python
marzo23 = pd.read_excel('/Users/martincarrasco/Desktop/Martín_Carrasco/Gestión PADS y Brand ADS/PADS MENSUALES/PADS MENSUALES 2025-07.xlsx', sheet_name = 'MARZO 23')

marzo23 = marzo23.dropna(how = 'all').dropna(how = 'all', axis = 1)

marzo23
```

```python
metricas = marzo23.iloc[::2].reset_index(drop=True)
valores = marzo23.iloc[1::2].reset_index(drop=True)
```

```python
print(metricas.columns)
print(valores.columns)
```

```python
valores['MÉTRICA'] = metricas['TOTALES']
valores.rename(columns = {'TOTALES': 'CONSOLIDADO'}, inplace = True)
valores
```

```python
valores_long = valores.melt(
    id_vars="MÉTRICA",
    var_name="Cuenta Meli",
    value_name="Valor"
)
```

```python
valores_final = valores_long.pivot(
    index="Cuenta Meli",
    columns="MÉTRICA",
    values="Valor"
).reset_index()

valores_final.columns.name = None
valores_final
```
